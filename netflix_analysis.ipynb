```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/netflix_titles.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.dropna(inplace = True)
```

```python
top_release = df['release_year'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_release.index, y=top_release.values)
plt.title("Top 10 Movies Release Year with Most Titles on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
rating_counts = df['rating'].value_counts().head(6)
plt.figure(figsize=(7,7))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Netflix Content Ratings")
plt.show()
```

```python
rating_type = pd.crosstab(df['cast'], df['type'])

plt.figure(figsize=(10,6))
sns.heatmap(rating_type, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of  vs Type")
plt.show()
```

```python
year_type = pd.crosstab(df['duration'], df['type'])

plt.figure(figsize=(12,10))
sns.heatmap(year_type, cmap="Blues", cbar=True)
plt.title("Heatmap of Content by Duration Vs Type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="rating", y="release_year", hue="type")
plt.title("Scatter Plot of Rating vs  Release Year")
plt.show()
```
